```python
def remove_skill_columns(df):
    skill_columns = [col for col in df.columns if 'Skill' in col]
    df.drop(axis=1, columns=skill_columns, inplace=True)
    return df

gps = remove_skill_columns(gps)
rps = remove_skill_columns(rps)
display(rps)
```

```python
def change_column_names(df, label):
    c = [n for n in range(len(df.columns))]
    c[0] = label
    df.columns = c
    return df

gps = change_column_names(gps, 'g')
rps = change_column_names(rps, 't')
display(rps)
```

```python
def get_mappings(df):
    
    # create numerical values to map to
    mapping = pd.Series(df.iloc[:,0].unique())
    mapping = mapping.to_dict()
    
    # flip dictionary
    inv_mapping = {v: k for k, v in mapping.items()}
    
    return inv_mapping
```

```python
# Returns true if grad 'g' prefers team 'r1' (current) over team 'r' (new)
def gPrefersR1OverR(gps, rps, g, r, r1):
    
    # check if g prefers r over their current team r1
    for i in range(len(gps[g])):
        
        # if r1 comes before r in g's preferences, do nothing
        if (gps[g][i] == r1):
            return True
        
        # if r comes before r1 in g's preferences, update g's match
        if (gps[g][i] == r):
            return False
        
    return False
```
